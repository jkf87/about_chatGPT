```python
import openai

openai.api_key = "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"

def chatbot(text):
    messages = []
    # 파일에서 데이터를 읽어들입니다.
    with open('pdf2text.txt', 'r', encoding='UTF-8') as f:
        data = f.read()

    # 데이터를 청크별로 챗봇 모델에 입력합니다.
    chunk_size = 4096
    chunks = [data[i:i+chunk_size] for i in range(0, len(data), chunk_size)]

    keepgoing = "I'll split up the entire text of the academic journal paper. Your task is to respond only with 'Okay' until I type 'Complete' first. When I type 'Complete' you will follow the steps below."
    messages = [{"role": "system", "content": keepgoing}]
    for chunk in chunks:
        messages = []  # 메시지 목록을 초기화합니다.
        messages.append({"role": "user", "content": chunk})
        chat = openai.Completion.create(
            engine="text-davinci-002",
            prompt=messages,
            max_tokens=1024,
            n=1,
            stop=None,
            temperature=0.7,
        )
        reply = chat.choices[0].text
        print(f"ChatGPT: {reply}")
        messages.append({"role": "assistant", "content": reply})

    # "Complete"가 입력될 때까지 대화를 이어나갑니다.
    while True:
        message = text
        if message.lower() == "complete":
            break
        elif message:
            messages = []  # 메시지 목록을 초기화합니다.
            messages.append({"role": "user", "content": message})
            chat = openai.Completion.create(
                engine="text-davinci-002",
                prompt=messages,
                max_tokens=1024,
                n=1,
                stop=None,
                temperature=0.7,
            )
            reply = chat.choices[0].text
            print(f"ChatGPT: {reply}")
            messages.append({"role": "assistant", "content": reply})

    return reply
```
